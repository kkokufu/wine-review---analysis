## (1) The variety of wine recommended most

```python
df_95_30 = df_95_36[df_95_36['price']<30].sort_values('price',ascending=True)
df_95_30
```

```python
df_95_36[df_95_36['price']<30].sort_values('points',ascending=False)
# priceが３０未満かつpointsが９５以上の条件を満たしたもののpointsが高い順に並びかえる。
```

```python
df2[(df2['points']==96) & (df2['price']==20)]
# points<=95,price<30を満たすワインの中で、最もpointsが高いワインはpoints=96であり、最もpriceが低いワインはprice=20なのでその両方の条件を満たすワインが存在するか調べる。
```

## (2)countries I would recommend Hotel TULIP to source wine from

```python
tmp = df2.groupby(['country', 'variety']).agg({'price':['mean'], 'points':['mean'], 'variety':'count'}).reset_index()
tmp
# ワインの種類のごとに原産国、priceの平均、pointsの平均、ワインの種類のカウント数の属性を含むテーブルを作成。 
```

```python
df_country = tmp.groupby('country').apply(lambda d: d.loc[d[('variety','count')].idxmax()])
df_country
# 更に、それぞれの国ごとに最もレビューされている種類のワインだけに制限してテーブルを作成。
```

```python
df_dropcountry = df_country.drop(columns=['country'])
df_dropcountry
# 原産国を表す列が二つ存在するので片方を削除
```

```python
df3 = df_dropcountry[(df_dropcountry['price']['mean']<=36) & (df_dropcountry['points']['mean']>=89)]
df3
#  df2.describe()より、"price","points"の平均値がそれぞれ35,88点代なので、よいワインの条件をprice<=36,points>=89と設定し、indexを絞り込む
```

```python
#上の表の６種類の国が原産国のワインを調達することをお勧めし、表のワインの種類がそれぞれの国で最もお勧めできるワインである。さらに、”variety count”を考慮すると圧倒的にAustriaのワインがレビューの数が多いので上の表ではAustriaのワインが最も信頼性が高いと言える。
```
